Week 2 Practical Lab : Inventory Analysis (This Project Combines the use of lists, Dicts and Loops)
Scenario: You are a data analyst for a retail store , you have a list of products that you must audit.

Requirements:1. Create a List of Dictionaries 
2 Loop through the List 
3 Use if Statement
4 Print the total value of the store inventory


```python
# Step 1: Create list of dictionaries

products = [
    {"name": "Rice", "price": 400, "quantity": 10},
    {"name": "Beans", "price": 240, "quantity": 3},
    {"name": "Milk", "price": 130, "quantity": 8},
    {"name": "Bread", "price": 120, "quantity": 2},
    {"name": "Eggs", "price": 100, "quantity": 12}
]
```

```python
# Step 2: Loop through the list
total_items_value = 0

for product in products:
    # Calculate total value for each product
    total_value = product["price"] * product["quantity"]
    total_items_value += total_value
    # Check low stock
    if product["quantity"] < 5:
        print(f"ALERT: {product['name']} is low on stock!") 
    print(f"{product['name']} total value = {total_value}")
```

```python
# Step 3: Final summary
print("\nTotal Inventory Value =", total_items_value)
```

TASK 2
Scenario: You have a database of 10 Employees 
Requirement 1: Write a loop to identify employees who earn more than 300,000 usd and work in the data department
Requirement 2: Create a new list called High Earners and append their names to it during the loop.
Requirement 3: if an employee's salary is missing, use the continue keyword to skip them and print a warning 
Requirement 4: Calculate the average salary of the filtered list only.

```python
# Sample dataset (List of Dictionaries)
employees = [
    {"name": "Prestige", "department": "Data", "salary": 350000},
    {"name": "Bob", "department": "HR", "salary": 250000},
    {"name": "Enoch", "department": "Data", "salary": None},
    {"name": "David", "department": "Data", "salary": 500000},
    {"name": "Opeyemi", "department": "Finance", "salary": 450000},
    {"name": "Frank", "department": "Data", "salary": 280000},
    {"name": "Grace", "department": "Data", "salary": 320000},
    {"name": "Iremide", "department": "Marketing", "salary": None},
    {"name": "Ivy", "department": "Engineering", "salary": 600000},
    {"name": "Jack", "department": "Data", "salary": 150000}
]



# Requirement 2: list for high earners
high_earners = []

# To calculate average
total_salary = 0
count = 0

# Loop through employees
for emp in employees:
    name = emp["name"]
    dept = emp["department"]
    salary = emp["salary"]
# Requirement 3: Handle missing salary
    if salary is None:
       print(f"Data Integrity Error for Employee {name}")
       continue
    # Requirement 1: Filter condition
    if salary > 300000 and dept == "Data":
        high_earners.append(name)
        total_salary += salary
        count += 1

# Requirement 4: Calculate average salary
average_salary = total_salary / count if count > 0 else 0

# Output results
print("\nHigh Earners:", high_earners)
print(f"Average Salary of High Earners: ${average_salary:,.0f}")
     
```
